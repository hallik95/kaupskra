```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vextir = pd.read_csv('bankavextirogdrattarvextir.csv') 
```

```python
vextir.head(20)
```

```python
vex = vextir.iloc[8:]
vex.columns = vex.iloc[0]
vex = vex[2:254] # Breytti hér því þetta endaði á einhverju sem við viljum ekki
vex.tail()
```

```python
df = vex[['Almennir vextir óverðtryggðra lána 4', 'Almennir vextir verð-tryggðra lána 4']]
df.tail()
```

```python
df = vex.iloc[:, [0,1,11,12]]
df.head()
```

```python
df = df.iloc[39:]
df.head()
```

```python
df = df.rename(columns = {'Almennir vextir óverðtryggðra lána 4':'overdtryggt', 'Almennir vextir verð-tryggðra lána 4':'verdtryggt'})
```

```python
df.columns.values[0] = 'ar'
df.columns.values[1] = 'man'
df.head()
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['ar'].ffill(axis=0, inplace=True)

```

```python
df.head()
```

```python
df.isna().sum() 
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df['man'] = pd.factorize(df['man'])[0]+1
df.head(20)
```

```python
df['man-ar'] = df['man'].astype(str) + '/' + df['ar'].astype(str)
df.head()
```

```python
df['man-ar'] = pd.to_datetime(df['man-ar'])
```

```python
df.dtypes
```

```python
df.head()
```

```python
vextir = df[['man-ar', 'overdtryggt', 'verdtryggt']]
```

```python
vextir.head()
```
